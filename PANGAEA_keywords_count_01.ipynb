### Read the Portal of Experimental Ocean Acidification Data

See python notebook: PANGAEA_Harvesting_OAI-PMH.ipynb

```python
import pandas as pd
```

```python
df = pd.read_csv("metadata.csv")
df['keywords'] = df['keywords'].apply(lambda x: eval(x))
df['authors'] = df['authors'].apply(lambda x: eval(x))
df['year'] = pd.to_datetime(df['year'], format='%Y')
df
```

```python
# Extract all keywords
allKeywordsDB = sorted(df['keywords'].apply(pd.Series).stack().unique())
print(len(allKeywordsDB))
allKeywordsDB
```

```python
# Find number of datasets for each keyword for the 2 periods

dfreq = pd.DataFrame()
year = 2015

for k in allKeywordsDB:
    s_total = df[df['keywords'].apply(lambda x: bool(set(x) & set([k]) if x != None else []))]
    s_period1 = s_total[s_total['year'].between('1950-01-01', str(year-1)+'-12-31')]
    s_period2 = s_total[s_total['year'].between(str(year)+'-01-01', '2030-12-31')]
    dfreq = dfreq._append({ 'keyword': k, 
                    'before ' + str(year): len(s_period1), 
                    'after ' + str(year):  len(s_period2)}, 
                    ignore_index=True)
```

```python
# Specify int type

dfreq = dfreq.astype({ 'before ' + str(year): int, 
            'after ' + str(year):  int
          })
```

```python
# Save to CSV file

dfreq.to_csv('OA-ICC_keywords_before_after_2015.csv', index=False)
```
