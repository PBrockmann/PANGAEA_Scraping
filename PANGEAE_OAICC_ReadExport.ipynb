### Read exported bibliodatabase files

```python
import pandas as pd
import csv
```

```python
tagDate = '20211203'
```

```python
dfArray = []
```

```python
file = 'OAICCincomplete_' + tagDate + '.utf8'

df = pd.read_csv(file, sep=';', encoding='utf_8', names=['Reference', 'DOI', 'Country', 'Year'])
df['Reference'] = df['Reference'].str.strip()
df['DOI'] = df['DOI'].str.strip()
df['Country'] = df['Country'].str.strip()
df['Year'] = df['Year']
df['Included/Not included'] = 'Not included'
df['Reason for non inclusion'] = '< 2 carbonate chemistry parameters'

dfArray.append(df)

df.head()
```

```python
file = 'OAICCnoanswer_' + tagDate + '.utf8'

df = pd.read_csv(file, sep=';', encoding='utf_8', names=['Reference', 'DOI', 'Country', 'Year'])
df['Reference'] = df['Reference'].str.strip()
df['DOI'] = df['DOI'].str.strip()
df['Country'] = df['Country'].str.strip()
df['Year'] = df['Year']
df['Included/Not included'] = 'Not included'
df['Reason for non inclusion'] = 'No answer from authors'

dfArray.append(df)

df.head()
```

```python
file = 'OAICCdatalost_' + tagDate + '.utf8'

df = pd.read_csv(file, sep=';', encoding='utf_8', names=['Reference', 'DOI', 'Country', 'Year'])
df['Reference'] = df['Reference'].str.strip()
df['DOI'] = df['DOI'].str.strip()
df['Country'] = df['Country'].str.strip()
df['Year'] = df['Year']
df['Included/Not included'] = 'Not included'
df['Reason for non inclusion'] = 'Data lost'

dfArray.append(df)

df.head()
```

```python
file = 'OAICCdb_' + tagDate + '.utf8'

df = pd.read_csv(file, sep=';', encoding='utf_8', names=['Reference', 'DOI', 'DOI_PANGAEA', 'Country', 'Year'])
df['Reference'] = df['Reference'].str.strip()
df['DOI'] = df['DOI'].str.strip()
df['DOI_PANGAEA'] = df['DOI_PANGAEA'].str.strip()
df['Country'] = df['Country'].str.strip()
df['Year'] = df['Year']
df['Included/Not included'] = 'Included'
df['Reason for non inclusion'] = ''

dfArray.append(df)

df.head()
```

```python
df = pd.concat(dfArray, sort=False)
print(len(df))

df[['Reference', 'DOI', 'DOI_PANGAEA', 'Included/Not included', 'Reason for non inclusion', 'Country', 'Year']]\
    .to_csv('OAICC_' + tagDate + '.csv', index=False, encoding='utf_8', quoting=csv.QUOTE_ALL)

```
